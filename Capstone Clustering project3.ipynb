Transform wikipedia webpage into dataframe

```python
data = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
# webscrape wikipedia page using beautifulsoup object
soup = BeautifulSoup(data, 'html.parser')
postalCodeList = []
boroughList = []
neighborhoodList = []
soup.find('table').find_all('tr')
soup.find('table').find_all('tr')
for row in soup.find('table').find_all('tr'):
    cells = row.find_all('td')
    if(len(cells) > 0):
        postalCodeList.append(cells[0].text)
        boroughList.append(cells[1].text)
        neighborhoodList.append(cells[2].text.rstrip('\n'))
```

```python
df = pd.DataFrame({"PostalCode": postalCodeList,
                   "Borough": boroughList,
                   "Neighborhood": neighborhoodList})

df.head()
```

Remove cells with borough as Not assigned

```python
df_dropna = df[df.Borough != "Not assigned"]
df_dropna.head()
```

Combine neighbourhoods

```python
df_com = df_dropna.groupby(["PostalCode", "Borough"], as_index=False).agg(lambda x: ", ".join(x))
df_com.head()
```

If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough

```python
for index, row in df_com.iterrows():
    if row["Neighborhood"] == "Not assigned":
        row["Neighborhood"] = row["Borough"]
        
df_com.head()
```

```python
radius = 500
LIMIT = 100

venues = []

for lat,long,post,borough,neighborhood in zip(df_new['Latitude'],df_new['Longitude'],df_new['PostalCode'],df_new['Borough'],df_new['Neighborhood']):
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(CLIENT_ID,CLIENT_SECRET,
        VERSION,lat,long,radius, LIMIT)
    results = requests.get(url).json()["response"]['groups'][0]['items']
    for venue in results:
        venues.append((post,borough,neighborhood,lat,long,venue['venue']['name'],venue['venue']['location']['lat'], venue['venue']['location']['lng'],
        venue['venue']['categories'][0]['name']))
```

```python
venues_df = pd.DataFrame(venues)
venues_df.columns = ['PostalCode', 'Borough', 'Neighborhood', 'BoroughLatitude', 'BoroughLongitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']
print(venues_df.shape)
venues_df.head()
```

```python
venues_df.groupby(["PostalCode", "Borough", "Neighborhood"]).count()
```

```python
toronto_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

toronto_onehot['PostalCode'] = venues_df['PostalCode'] 
toronto_onehot['Borough'] = venues_df['Borough'] 
toronto_onehot['Neighborhoods'] = venues_df['Neighborhood'] 

fixed_columns = list(toronto_onehot.columns[-3:]) + list(toronto_onehot.columns[:-3])
toronto_onehot = toronto_onehot[fixed_columns]

print(toronto_onehot.shape)
toronto_onehot.head()
```

```python
toronto_com = toronto_onehot.groupby(["PostalCode", "Borough", "Neighborhoods"]).mean().reset_index()

print(toronto_com.shape)
toronto_com
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

areaColumns = ['PostalCode', 'Borough', 'Neighborhoods']
freqColumns = []
for ind in np.arange(num_top_venues):
    try:
        freqColumns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        freqColumns.append('{}th Most Common Venue'.format(ind+1))
        
columns = areaColumns+freqColumns

neighborhoods_venues = pd.DataFrame(columns=columns)
neighborhoods_venues['PostalCode'] = toronto_com['PostalCode']
neighborhoods_venues['Borough'] = toronto_com['Borough']
neighborhoods_venues['Neighborhoods'] = toronto_com['Neighborhoods']

for ind in np.arange(toronto_com.shape[0]):
    row_categories = toronto_com.iloc[ind, :].iloc[3:]
    row_categories_sort = row_categories.sort_values(ascending=False)
    neighborhoods_venues.iloc[ind, 3:] = row_categories_sort.index.values[0:num_top_venues]

print(neighborhoods_venues.shape)
neighborhoods_venues
```

```python
kclusters = 3

toronto_clustering = toronto_com.drop(["PostalCode", "Borough", "Neighborhoods"], 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_clustering)

kmeans.labels_[0:10]
```

```python
toronto_grouped = df_new.copy()
toronto_grouped["Cluster Labels"] = kmeans.labels_
toronto_grouped = toronto_grouped.join(neighborhoods_venues.drop(["Borough", "Neighborhoods"], 1).set_index("PostalCode"), on="PostalCode")
print(toronto_grouped.shape)
toronto_grouped.head()
```

```python
toronto_grouped.sort_values(["Cluster Labels"], inplace=True)
toronto_grouped
```
